```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## ZORI

```python
zori = pd.read_csv('./data/cleandata/clean_zori.csv').drop('Unnamed: 0', axis = 1)
```

```python
print(f"Data contains {pd.unique(zori['ZipCode']).shape[0]} unique zipcodes.")
print(f"Data contains {pd.unique(zori['County'] + ', ' + zori['State']).shape[0]} unique counties.")
```

```python
sns.distplot(zori['ZORI']); #Skewed right, i.e., a lot of outliers
```

```python
sns.distplot(zori[zori['ZORI'] <= 4000]['ZORI']);
```

```python
print(f"Out of {zori.shape[0]} ZORI rows, {zori[zori['ZORI'] <= 4000].shape[0]} were less than $4,000.")
```

```python
temp = zori[zori['ZORI'] <= 4000].groupby(['Year', 'Month']).mean()['ZORI'].reset_index()
temp['Year_Month'] = temp['Year'].map(str) + '_' + temp['Month'].map(str)
temp['Year_Month'] = pd.to_datetime(temp['Year_Month'], format = '%Y_%m')
temp = temp[['Year_Month', 'ZORI']]
plt.figure(figsize = (12, 6))
plt.plot(temp['Year_Month'], temp['ZORI'])
plt.xticks(rotation = 90)
plt.title('Trend of national average ZORI over time');
```

## Income and Household

```python
inc_hh = pd.read_csv(r"C:\Users\Steven Lantigua\Desktop\data\income_and_households.csv").drop('Unnamed: 0', axis = 1)
inc_hh['Location'] = inc_hh.County +', '+ inc_hh.State
```

```python
inc_hh.info()
```

### Distibution of median income and number of households

```python
sns.distplot(inc_hh['Med_income'])
```

```python
sns.distplot(inc_hh['Total_Households'])
```

Both features are skewed to the right indicating existing outliers that can distort data analysis and affect model performance 

### *top 10 counties with highest average household median income*

```python
len(inc_hh.County.unique())
```

```python
len(inc_hh.Location.unique())
```

```python
inc_hh.groupby('Location')['Med_income'].mean().nlargest(10).plot.bar(title='Highest Average Household Median Income ')
```

top 10 counties with the highest average household median income between 2012 & 2019

### *Fastest Growing Counties by percent change in total households*

```python
inc_hh.groupby('Location')['Total_Households'].pct_change().nlargest(10).plot(kind='bar')
```

```python
inc_hh.groupby('Location')['Total_Households'].mean()
```

```python
inc_hh['pct_ch'] = (inc_hh.groupby(['Location','Year'])['Total_Households']
                                  .apply(pd.Series.pct_change) + 1)
```

```python
inc_hh.groupby('Location')['pct_ch'].mean().nlargest(10)
```

```python
inc_hh.groupby('County')['pct_ch'].mean().nlargest(10).plot.bar(title='Fastest Growing Counties (percent change)')
```

Top 10 counties with the highest average percent change in total households between 2012 and 2019

```python
cts = inc_hh.groupby('County')['pct_ch'].mean().nlargest(10).index.to_list()
bycounty = inc_hh.set_index('County').loc[cts]
bycounty.groupby('County')['Med_income'].mean().sort_values()
```

almost all of the counties are below the national median household income ($62,843)

- Wayne County:  is the most populous county in the U.S. state of Michigan.
- DeKalb County: As of the 2010 census, the population was 691,893, making it Georgia's fourth-most populous county. It's included in the Atlanta-Sandy Springs-Roswell, GA Metropolitan Statistical Area.

## Job openings

```python
job_opns = pd.read_csv(r"C:\Users\Steven Lantigua\Desktop\7ParkCP\clean_jos.csv").drop('Unnamed: 0', axis = 1)
```

```python
job_opns.info()
```

### States average job openings rate 

```python
plt.subplot(1, 2, 1)
job_opns.groupby('State')['Job Openings Rate'].mean().nlargest(5).plot.bar(title='Top 5 Job Openings Rate')
plt.subplot(1, 2, 2)
job_opns.groupby('State')['Job Openings Rate'].mean().nsmallest(5).plot.bar(title='Bottom 5 Job Openings Rate')
plt.ylim((0,6))
plt.show()
```

- Top 5: Alaska, Montana, Wyoming, North Dakota, and Virginia
- Bottom 5: Connecticut, DC, New York, New Jersey, Mississipi 

### States average hire rate 

```python
plt.subplot(1, 2, 1)
job_opns.groupby('State')['Hires Rate'].mean().nlargest(5).plot.bar(title='Top 5 Hires Rate by State')
plt.subplot(1, 2, 2)
job_opns.groupby('State')['Hires Rate'].mean().nsmallest(5).plot.bar(title='Bottom 5 Hires Rate by State')
plt.ylim((0,6))
plt.show()
```

## Vacancy

```python
vac = pd.read_csv("..\data\cleandata\clean_vacancy.csv").drop('Unnamed: 0', axis=1)
```

```python
vac['Location'] = vac['County'] + ', ' + vac['State']
```

```python
vac.info()
```

```python
sns.distplot(vac['Rental Vacancy Rate'])
```

### Yearly Median Vacancy Rate 

```python
vac.groupby('Year')['Rental Vacancy Rate'].median().plot(kind='bar',legend=False, title='Yearly Vacancy Rate')
plt.ylabel('Vacancy Rate (%)')
plt.show()
```

```python
vac.groupby('Year')['Rental Vacancy Rate'].median().plot(kind='line',legend=False, title='National Trend of Vacancy Rate')
plt.ylabel('Vacancy Rate')
plt.show()
```

- Overall, rental vacancy rate seems to be going down over the years. Housing market started to recover around 2010.

### Vacancy rate by Location

```python
vac.groupby('Location')['Rental Vacancy Rate'].median().nsmallest(10).plot(kind='bar', title='Top 10 Counties with Lowest Vacancy Rate')
plt.ylabel('Vacancy Rate (%)')
plt.show()
```

```python
vac.groupby('Location')['Rental Vacancy Rate'].median().nsmallest(10)
```

### Vacancy rate by state

```python
vac.groupby('State')['Rental Vacancy Rate'].median().nsmallest(10).plot(kind='bar',title='Top 10 States with Lowest Vacancy Rate')
plt.ylabel('Vacancy Rate (%)')
plt.show()
```

```python
vac.groupby('State')['Rental Vacancy Rate'].median().nsmallest(10)
```

- using median took mitigate distortion done by outliers
